```python
import numpy as np
import xarray as xr
import pandas as pd
import metpy.calc as mpcalc
```

```python
#read the data

IAN_ds = xr.open_dataset('HurricaneIAN.nc')
```

```python
IAN_ds #view data
```

```python
#IAN_ds.keys()
IAN_ds.time
```

```python
# The data is in 2 hour intervals

# Let's convert to days
IAN_ds_daily = IAN_ds.resample(time = 'D').mean()
```

```python
IAN_ds_daily
```

```python
#Converting to monthly
IAN_ds_monthly = IAN_ds.resample(time = 'ME').mean() #Month start MS. Month end ME
```

```python
IAN_ds_monthly
```

```python
#Hourly: H, Minutes: min, Yearly: Y

IAN_ds_4hourly = IAN_ds.resample(time = '4H').mean()
```

```python
IAN_ds_4hourly
```

```python
# Group by operation for seasonal mean

IAN_seasonal_ds = IAN_ds.groupby('time.season').mean()
```

```python
IAN_seasonal_ds
```

```python
IAN_diurnal = IAN_ds.groupby('time.hour').mean()
```

```python
IAN_diurnal
```

```python
#Save the data via daily mean temperature

#1. select tmp from fata
#2. convert 2 hourly data
#3. Covert to CSV
#4. we need to save csv

# IAN_tmp = ((IAN_ds.t - 273.15)*9/5) + 32
IAN_tmp = IAN_ds.t
```

```python
IAN_tmp
```

```python
daily_tmp = IAN_tmp.resample(time = 'D').mean() 
```

```python
#convert to table
IAN_tmp_table = daily_tmp.to_dataframe()
```
