```python
import tropycal.tracks as tracks
import datetime as dt
import matplotlib.pyplot as plt
```

```python
basin = tracks.TrackDataset(basin='north_atlantic',source='hurdat',include_btk=False)
```

```python
storm = basin.get_storm(('Irma',2017))
```

```python
storm.plot()

cities = {
    "Key West": (-81.7799, 24.5551),
}

for city, (lon, lat) in cities.items():
    plt.scatter(lon, lat, s=50, color = 'black', marker="o", edgecolor="black")  # City marker
    plt.text(lon, lat, city, fontsize=12, ha="right", va="bottom", color="Blue")  # City name

plt.savefig('Irma_Track.png',dpi=150, bbox_inches = 'tight')
```
