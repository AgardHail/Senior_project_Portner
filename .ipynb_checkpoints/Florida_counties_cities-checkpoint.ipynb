```python
import geopandas as gpd
import matplotlib.pyplot as plt
import requests
import zipfile
import io

# URL for Florida counties from the Census Bureau
url = "https://www2.census.gov/geo/tiger/TIGER2023/COUNTY/tl_2023_us_county.zip"

# Download and extract in memory
response = requests.get(url)
with zipfile.ZipFile(io.BytesIO(response.content)) as z:
    z.extractall("fl_counties")

# Read shapefile (filter for Florida, STATEFP=12)
shp_path = "fl_counties/tl_2023_us_county.shp"
counties = gpd.read_file(shp_path)
florida_counties = counties[counties["STATEFP"] == "12"]  # STATEFP=12 is Florida

# Define city coordinates (longitude, latitude) 
cities = {
    #"Melbourne": (-80.6081, 28.0836),
    #"Cape Canaveral": (-80.6043, 28.3922),
    #"Miami": (-80.1918, 25.7617),
    "Key West": (-81.7799, 24.5551),
    #"Pensacola": (-87.2169, 30.4213),
    #"Jacksonville": (-81.6557, 30.3322),
    #"Tampa": (-82.4572, 27.9506),  # Added Tampa!
}

# Plot Florida counties
fig, ax = plt.subplots(figsize=(12, 12))
florida_counties.plot(ax=ax, color='lightblue', edgecolor='black')

# Add county labels
for _, row in florida_counties.iterrows():
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, row["NAME"], 
             fontsize=8, ha="center", color="darkblue")

# Add city labels
for city, (lon, lat) in cities.items():
    plt.scatter(lon, lat, color='red', s=50, marker="o", edgecolor="black")  # City marker
    plt.text(lon, lat, city, fontsize=10, ha="left", va="bottom", color="red")  # City name

# Labels and title
ax.set_xlabel("Longitude", fontsize=12)
ax.set_ylabel("Latitude", fontsize=12)
ax.set_title("Florida Counties Map with City Labels", fontsize=14)

# Show the map
plt.show()
#plt.savefig('Florida_Counties_map.png',dpi=150)
```
